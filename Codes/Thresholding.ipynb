## Load Libraries

```python
import skimage
import matplotlib.pyplot as plt
```

## Load Image

```python
image = plt.imread("flag_image.jpg")
plt.axis('off')
plt.imshow(image);
```

## Convert img to Grayscale

```python
gray_image = skimage.color.rgb2gray(image)
plt.axis('off')
plt.imshow(gray_image, cmap='gray');
```

## Apply Thresholding

#### **Choosing the right threshold**

```python
plt.hist(gray_image.ravel(), bins=256);
```

```python
# As the peak is 1.0 Manually choose a
# threshold slightly below the peak. Adjust as needed
thresh = 0.60
```

```python
# Create a binary image where pixels above the threshold
# become True (white), and pixels below or equal become False (black)
binary = gray_image > thresh


# Explaination:
# The expression gray_image > thresh creates a boolean array where each
# element is True if the corresponding pixel intensity in gray_image is
# greater than thresh, and False otherwise. This boolean array is then
# assigned to the variable binary.

# When this boolean array is used for visualization
# (e.g., with plt.imshow(binary, cmap='gray')),
# Matplotlib interprets True as white and False as black.
# Therefore, pixels with intensities greater than the
# threshold will be displayed as white in the binary image.

# In summary, the visualization convention is built into the
# way boolean arrays are interpreted by Matplotlib when used
# with the imshow function. True values are treated as white,
# and False values are treated as black by default.
```

```python
plt.imshow(binary, cmap='gray');
```
